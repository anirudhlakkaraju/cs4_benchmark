```python
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
import pandas as pd
```

```python
path = "Path to Writing Prompts dataset - train.wp_target"
```

```python
with open(path) as f:
  stories = f.readlines()
stories = [" ".join(i.split()[0:1000]) for i in stories]
print(len(stories))
```

```python
# Convert the list into a DataFrame
df = pd.DataFrame(stories, columns=["Stories"])
df.head()
```

```python
# Print story and check
stories[0]
```

```python
# Compute length of each story
def count_words_excluding_newline(text):
    # Tokenize by spaces and filter out "<newline>"
    words = [word for word in text.split() if word != "<newline>"]
    return len(words)

df['Length'] = df['Stories'].apply(count_words_excluding_newline)
df.head()
```

```python
# Sort stories by length in descending order 
df = df.sort_values(by='Length', ascending=False)
```

```python
# Select the top 60 rows
new_dataset = df.head(60)

# Save the new DataFrame to a CSV file
new_dataset.to_csv('new_dataset.csv', index=False)
files.download('new_dataset.csv')
```
